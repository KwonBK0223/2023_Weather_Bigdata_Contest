```python
import pandas as pd
import numpy as np
from haversine import haversine
```

```python
total = pd.read_csv("../기상청공모전/data/total_time_gap.csv")
total.head()
```

```python
for i in range(1, len(total)):
    total.loc[i, 'km/s'] = haversine((total.loc[i, 'latitude'],total.loc[i, 'longitude']),(total.loc[i-1, 'latitude'],total.loc[i-1, 'longitude']),unit = 'km') / total.loc[i, 'time_gap']
```

```python
for i in range(2, len(total)):
    total.loc[i, 'km/s_2'] = (total.loc[i, 'km/s'] - total.loc[i-1, 'km/s']) / total.loc[i,'time_gap']
```

```python
total
```

```python
# 각 배 번호의 1번째 2번째 삭제
ship_num_busan = total['num'].unique()

_drop_busan = []
for i in ship_num_busan:
    k = total[total["num"] == i].index[0]
    _drop_busan.append(k)
    k += 1
    _drop_busan.append(k)
    
total = total[~total.index.isin(_drop_busan)].reset_index(drop=True)
```

```python
total = total[['num', 'time', 'latitude', 'longitude', 'sog', 'cog', 'hdg', 'time_gap', 'km/s', 'km/s_2', 'anchor_drag']]
total.columns = ['num', 'time', 'latitude', 'longitude', 'sog', 'cog', 'hdg', 'time_gap', 'd_km/s', 'dd_km/s', 'anchor_drag']
```
