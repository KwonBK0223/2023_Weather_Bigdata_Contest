```python
# 도분초 -> 위도,경도 변환 함수
def to_dec(coordinate):
    temp = coordinate.split()
    temp = [float(i) for i in temp[1].split("-")]
    
    deg = temp[0]
    min = temp[1]
    sec = temp[2]
    
    temp1 = deg + min/60 + sec/3600
    return round(temp1,6)
```

```python
위도 = [to_dec("N 35-19-07"), to_dec("N 35-09-53"), to_dec("N 35-08-56"), to_dec("N 35-19-07"), to_dec("N 35-05-42"), 
      to_dec("N 35-17-50"), to_dec("N 35-01-21"), to_dec("N 35-22-01"), to_dec("N 35-10-57"), to_dec("N 35-34-40")]
```

```python
# mask 씌울 항목 선정 - latitude = 'N35.064-99' 인 항목은 float으로 바꿀 수 없으므로 제외함
mask1 =busan_anch_train_final['busan_anch_train_final.latitude'].isin(['N35.064-99'])
busan_anch_train_final = busan_anch_train_final[~mask1] # ~를 포함하게 되면 mask의 값을 제외, ~을 제외하면 mask의 값을 포함

# mask 씌울 항목 선정 - longitude = 'E129.053-9' 인 항목은 float으로 바꿀 수 없으므로 제외함
mask2 =busan_anch_train_final['busan_anch_train_final.longitude'].isin(['E129.053-9'])
busan_anch_train_final = busan_anch_train_final[~mask2] # ~를 포함하게 되면 mask의 값을 제외, ~을 제외하면 mask의 값을 포함

busan_anch_train_final = busan_anch_train_final.replace(regex=r'[N]', value='') # latitude 데이터의 'N'삭제
busan_anch_train_final = busan_anch_train_final.replace(regex=r'[E]', value='') # longititude 데이터의 'E'삭제


busan_anch_train_final['busan_anch_train_final.latitude'] = busan_anch_train_final['busan_anch_train_final.latitude'].astype(float) # 모든 데이터 타입을 float으로 변경
busan_anch_train_final['busan_anch_train_final.longitude'] = busan_anch_train_final['busan_anch_train_final.longitude'].astype(float)
busan_anch_train_final['busan_anch_train_final.num'] = busan_anch_train_final['busan_anch_train_final.num'].astype(int) # num 데이터 타입을 int으로 변경
busan_anch_train_final.info()
```

```python
from scipy.spatial.distance import cdist

# 최근접 지점명을 반환하는 함수
def nearest_location(row):
    distances = cdist([(row['busan_anch_train_final.latitude'], row['busan_anch_train_final.longitude'])], 
                      beacon_buoy[['위도', '경도']], metric='euclidean')
    nearest_idx = distances.argmin()
    return beacon_buoy.loc[nearest_idx, '지점명']
```

```python
busan_anch_train = busan_anch_train_final.copy()
busan_anch_loc = busan_anch_train[['busan_anch_train_final.latitude','busan_anch_train_final.longitude']]
busan_anch_train['buoy_location'] = busan_anch_loc.apply(nearest_location, axis=1)
busan_anch_train
```

```python
tmp = beacon_buoy.copy()
tmp.rename(columns = {'지점명':'buoy_location'}, inplace = True)
busan_anch_train = pd.merge(busan_anch_train, tmp, how='inner')
busan_anch_train
```

```python
# 같은 번호를 하나로 묶고 buoy_location 값을 리스트로 모으는 함수
def merge_locations(group):
    locations = group['buoy_location'].unique().tolist()
    return pd.Series({'buoy_location': locations})

# 번호별로 buoy_location 값을 리스트로 모으기
merged_df = busan_anch_train.groupby('busan_anch_train_final.num').apply(merge_locations).reset_index()

# 결과 출력
merged_df
```
