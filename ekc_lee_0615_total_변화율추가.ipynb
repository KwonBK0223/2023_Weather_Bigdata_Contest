```python
import pandas as pd
import numpy as np
```

```python
total = pd.read_csv("../기상청공모전/data/total_data.csv")
total.head()
```

```python
total['time_gap'] = total['time']
```

```python
for i in range(1,len(total)):
    total['time_gap'][i] = (pd.to_datetime(total['time'][i]) - pd.to_datetime(total['time'][i-1])).total_seconds()
```

```python
total
```

```python
total1 = total
total1['time_gap'] = 0
```

```python
for i in range(1, len(total1)):
    if total1.loc[i, 'time_gap'] == 0:
        dt = (pd.to_datetime(total1.loc[i, 'time']) - pd.to_datetime(total1.loc[i-1, 'time'])).total_seconds()
        if dt != 0:
            total1.loc[i, 'time_gap'] = dt
        else:
            a = total1.loc[(total1['num'] == total1.loc[i, 'num']) & (total1['time'] == total1.loc[i, 'time'])].index.tolist()
            total1.loc[a, 'time_gap'] = 1 / len(a)

# 결과 출력
print(total1)
```
