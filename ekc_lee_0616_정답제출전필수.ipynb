```python
import pandas as pd
```

```python
df = pd.read_csv('../기상청공모전/data/0616.csv')
df
```

```python
busan = df[df['AREA']=='BUSAN']
ulsan = df[df['AREA']=='ULSAN']
```

```python
pred=[]
```

```python
for i in busan['NUM'].unique():
    ship = busan[busan['NUM']==i]
    cont = 0
    start = -1 
    for j in ship.index:
        if busan.loc[j,'TARGET'] == 1:
            if cont==0:
                start = j
            cont = cont + 1
        elif busan.loc[j,'TARGET'] == 0 and cont!=0:
            cont = 0
            start=-1
        if cont >= 5 :
            pred.append(start)
            break
            
```

```python
for i in ulsan['NUM'].unique():
    ship = ulsan[ulsan['NUM']==i]
    cont = 0
    start = -1 
    for j in ship.index:
        if ulsan.loc[j,'TARGET'] == 1:
            if cont==0:
                start = j
            cont = cont + 1
        elif ulsan.loc[j,'TARGET'] == 0 and cont!=0:
            cont = 0
            start = -1
        if cont >= 5 :
            pred.append(start)
            break
            
```

```python
print(len(pred))
```

```python
print(pred)
```

```python
df['TARGET'] = 0 
for i in pred:
    df.loc[i,'TARGET'] = 1
```
